```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Diwali.csv')

df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
df['Amount']=df['Amount'].fillna('0.0')
df['Marital_Status']=df['Marital_Status'].replace({0:'Unmarried',1:'Married'})
df.reset_index(drop=True, inplace=True)
df.head(10)

```

```python
# AGE V/S SALES 

gender={}
for row in df.Gender:
    gender[row]=gender.get(row,0)+1
    
genders=[]
count=[] 
for (key,value) in gender.items():
     genders.append(key)
     count.append(value)

bars=plt.bar(genders,count,color=['orange','green'],alpha=0.8)
plt.bar_label(bars, labels=count, label_type='edge')
plt.xlabel('Genders')
plt.ylabel('Counts')
plt.title('Gender V/S Purchasing')
plt.show()
    
```

```python
Marital_Statuses={}
for i in df.Marital_Status:
    Marital_Statuses[i]=Marital_Statuses.get(i,0)+1
Marital_Status=[]
Purchasing=[]
for key,value in Marital_Statuses.items():
    Marital_Status.append(key)
    Purchasing.append(value)

bars=plt.bar(Marital_Status,Purchasing,color=['#123740','#f1802d'],alpha=0.8)
plt.bar_label(bars, labels=Purchasing, label_type='edge')
plt.xlabel('Marital_Status')
plt.ylabel('Purchasing')
plt.title('Marital_Status V/S Purchasing')
plt.show()
    
```

```python


df['Product_Category']=df['Product_Category'].replace(['Footwear & Shoes','Beauty','Clothing & Apparel'],'Fashion')
df['Product_Category']=df['Product_Category'].replace(['Furniture','Decor','Household items','Tupperware'],'Household & Decor')
df['Product_Category']=df['Product_Category'].replace(['Auto','Hand & Power Tools'],'Industrial')
df['Product_Category']=df['Product_Category'].replace(['Stationery','Office','Veterinary','Books','Sports Products'],'Health & Education')
product_category={}
for row in df.Product_Category:
    product_category[row]=product_category.get(row,0)+1
Products=[]
Sales=[]
for key,value in product_category.items():
    Products.append(key)
    Sales.append(value)
plt.figure(figsize=(12, 10))
plt.pie(Sales, labels=Products,autopct='%.2f%%',startangle=140)
plt.show()
    



```

```python
Zone=np.array(df.Zone)
Order=np.array(df.Orders)
unique_zones = np.unique(Zone)
Zones=[]
Orders=[]
for zone in unique_zones:
    Zones.append(zone)
    Orders_by_zones=Order[Zone==zone]
    Orders.append(Orders_by_zones.sum())

bars=plt.bar(Zones,Orders,color=['#123740','#549aab','#b0d7e1','#cbcfd2','#f1802d'],alpha=0.8)
plt.bar_label(bars, labels=Orders, label_type='edge')
plt.xlabel('Zones')
plt.ylabel('Orders')
plt.title('Zones V/S Orders')
plt.show()
    
```

```python
df['State'].replace('AndhraPradesh', 'Andhra Pradesh')
State=np.array(df.State)
#np Array for orders is already defined
unique_States=np.unique(State)
States=[]
Orders=[]
for state in unique_States:
    Orders_by_states=Order[State==state]
    States.append(state)
    Orders.append(Orders_by_states.sum())

plt.figure(figsize=(8,8))
bars=plt.bar(States,Orders,color=['#123740','#549aab','#b0d7e1','#cbcfd2','#f1802d'],alpha=0.8)
plt.bar_label(bars, labels=Orders, label_type='edge')

plt.xticks(rotation=90)
plt.xlabel('Zones', fontsize=14, fontweight='semibold') 
plt.ylabel('Total Orders', fontsize=14, fontweight='semibold')  
plt.title('Total Orders by Zone', fontsize=16, fontweight='semibold') 
plt.show()

```
